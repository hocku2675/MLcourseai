```python
# for Jupyter-book, we copy data from GitHub, locally, to save Internet traffic,
# you can specify the data/ folder from the root of your cloned
# https://github.com/Yorko/mlcourse.ai repo, to save Internet traffic
DATA_URL = "https://raw.githubusercontent.com/Yorko/mlcourse.ai/main/data/"
```

```python
data = pd.read_csv(DATA_URL + "adult.data.csv")
data.head()
```

```python
#1
data["sex"].value_counts()
```

```python
#2
data[data["sex"] == "Female"]["age"].mean()
```

```python
#3
float((data["native-country"] == "Germany").sum()) / data.shape[0]
```

```python
#4-5
less_salary = data[data["salary"] == "<=50K"]["age"]
more_salary = data[data["salary"] == ">50K"]["age"]

print(less_salary.mean())
print(more_salary.mean())

print(less_salary.std())
print(more_salary.std())
```

```python
#6
high_salary = data[data["salary"] == ">50K"]
salary_after_high_education = high_salary["education"].isin(["Bachelors", "Prof-school", "Assoc-acdm", "Assoc-voc", "Masters or Doctorate"]).all()
salary_after_high_education
#2 способ
#data[data["salary"] == ">50K"]["education"].unique()  # No
```

```python
for (race,sex), sub_df in data.groupby(["race","sex"]):
    print(race,sex)
    print(sub_df["age"].describe())
```

```python
#8
data[(data["sex"] == "Male") 
    & (data["marital-status"].str.startswith("Married"))][
    "salary"
].value_counts(normalize=True)
```

```python
#9
data["hours-per-week"].max()
```

```python
person_count = data[data["hours-per-week"] == 99].shape[0]
person_count
```

```python
reach_percentage = float(data[(data["hours-per-week"] == 99) & (data["salary"] == ">50K")].shape[0]) / person_count
reach_percentage*100
```

```python
#10
for (native_country,salary), sub_df in data.groupby(["native-country","salary"]):
    print(native_country,salary, round(sub_df["hours-per-week"].mean()))
```
